In this group exercise, you will mimic what we did with Ï€: you will write code to calculate $e$ using the following series approximation (from [wikipedia](https://en.wikipedia.org/wiki/List_of_representations_of_e#As_an_infinite_series)):


$$ e = \sum\limits_{k=0}^{\infty} \frac{1}{k!} $$

where $k!$ represents 'k factorial': $k! = 1 \cdot 2 \cdot 3 \cdot \ldots \cdot k$


```python
""" import libraries """
# for storing series
import numpy as np

# for comparison
import math
```

```python
""" Calculate e using the series above """
# number of iterations
N = 100

# initialize e to 1
e = 1

# initialize the first factorial term to 1 as 0! = 1
k_factorial = 1   # no need to store all the factorials

# add up all the terms to update the estimate of e
for k in range(1, N):
  k_factorial *= k
  e += 1/k_factorial

# print e and compare the results
print("Calculated:", e)
print("Built-in:  ", math.e)
```

```python
""" Same calcualte with series stored for plotting """
# initialize the first factorial term back to 1 when k = 0
k_factorial = 1

# create an numpy array to store the series of e estimates
e_values = np.zeros(N)
e_values[0] = 1

# adding up the terms to approximate e
for k in np.arange(1, N):
  k_factorial *= k
  e_values[k] = e_values[k-1] + 1/k_factorial

print(e_values)
```

```python
# import libraries for plotting
import matplotlib.pyplot as plt
```

```python
""" Plot the convergence of e series """
plt.plot(e_values)

# plot a horizontal line indicating the true value of e
plt.axhline(math.e, color='gray', linestyle='--')

# display the plot
plt.show()
```
