This notebook is designed to be a live demonstration of Python fundamentals applied to calculating Ï€ using the [Madhava-Leibniz series](https://en.wikipedia.org/wiki/Approximations_of_%CF%80#Middle_Ages)

The approximation is simple:

$$ \pi \approx 4 \cdot \left(1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} \ldots -(-1)^n \frac{1}{2n - 1}  \right) $$

```python
""" Calculate pi based on Madhava-Leibniz series """
# text inbetween triple quotation marks indicates what the current cell is about

# initialize pi to 0
pi = 0

# better to set the number of terms to a variable rather than a fixed integer
N = 10000

# adding up terms to approximate pi
# no need to use numpy if just for simple iteration
for n in range(1, N):
  pi += -(-1)**n / (2*n-1)
pi = pi*4

# print pi in default precision
print(pi)
```

```python
""" Compare the answer against built-in value of pi """
# import libraries
import numpy as np
import math

# print the true value of pi
print(math.pi)
```

```python
""" Same calculation with converged series stored """
# create the numpy array to store pi series
pi_values = np.zeros(N)

# calcualte each term based on the previous one
for n in np.arange(1, N):
  pi_values[n] = pi_values[n-1] -(-1)**n / (2*n-1)
pi_values = pi_values*4

# print the pi series
print(pi_values)
```

```python
# import libraries for plotting
import matplotlib.pyplot as plt
```

```python
""" Plot the convergence of pi series """

plt.plot(pi_values[1:])

# set the scale of x-axis to log
plt.xscale('log')

# plot a horizontal line indicating the true value of pi
plt.axhline(math.pi, color='gray', linestyle='--')

# display the plot
plt.show()
```
